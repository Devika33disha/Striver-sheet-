```python
# 3. If Elseif
def studentGrade(marks):
  if marks > 90:
    print("A")
  elif marks > 70:
    print("B")
  elif marks > 50:
    print("C")
  elif marks > 35:
    print("D")
  else:
    print("Fail")
n = int(input())
studentGrade(n)
```

```python
# 4. Switch case
def WhichWeekDay(day):
  days = {
      1:"Monday",
      2:"Tuesday",
      3:"Wednesday",
      4:"Thursday",
      5:"Friday",
      6:"Saturday",
      7:"Sunday"
  }
  res = days.get(day, "Invalid Day") # either give the day or if its not in dict - then say invalid day
  print(res)
n = int(input())
WhichWeekDay(n)
```

```python
# 6.For Loop
def ForLoop(low , high):
  res = 0
  for i in range(low,high+1):
    res += i
  print(res)
low = int(input("Enter low:"))
high = int(input("Enter High:"))
ForLoop(low , high)
```

```python
# 7. While loop
def WhileLoop(n):
  count = 0
  total_sum = 0
  num = n if n != 0 else 10
  while count < 50:
    total_sum += num
    num += 10
    count += 1
  print(total_sum)
n = int(input("Enter number:"))
WhileLoop(n)
```

```python
# 8. Functions (Pass by referance and value)
def reverse(arr):
  n = len(arr)
  start = 0
  end = n-1
  while start < end:
    arr[start] , arr[end] = arr[end] , arr[start]
    start += 1
    end -= 1
  print(arr)
# taking array as input
l = int(input("Enter length of array:"))
arr = []
for i in range(l):
  arr.append(int(input(f"enter element {i+1}:")))

reverse(arr)
### another method
# l = int(input("Enter length of array:"))
# arr = []
# for i in range(l):
#   arr.append(int(input(f"enter element {i+1}:")))
# arr.reverse()
# print(arr)

```

```python
# pattern 3
def pattern3(n):
  for i in range(1,n+1):
    for j in range(1, i + 1):
        print(j, end="") # prints without space and without new line
    print() # for moving next line
n = int(input("Enter number:"))
pattern3(n)
```

```python
# pattern 7
def pattern7(n):
    for i in range(n):
      for j in range(n - i - 1):
          print(" ", end="")
      for k in range(2 * i + 1):
          print("*", end="")
      print()

n = int(input())
pattern7(n)
```

```python
# pattern 8
def pattern8(n):
  for i in range(n):
    for j in range(i):
      print(" " , end = "")
    for k in range(2* (n-i) - 1):
      print( "*" , end = "")
    print()
n = int(input("Enter number:"))
pattern8(n)
```

```python
# pattern 9
def pattern9(n):
    # top half - uper vaala diamond
    for i in range(n):
        for j in range(n - i - 1):
            print(" ", end="")
        for k in range(2 * i + 1):
            print("*", end="")
        print()

    # bottom half - neeche vaala diamond
    for i in range(n - 1, -1, -1):
        for j in range(n - i - 1):
            print(" ", end="")
        for k in range(2 * i + 1):
            print("*", end="")
        print()

n = int(input("Enter number:"))
pattern9(n)
```

```python
# pattern 10
def pattern10(n):
    for i in range(1, n + 1):
        for j in range(i):
            print("*", end="")
        print()
    for i in range(n - 1, 0, -1):
        for j in range(i):
            print("*", end="")
        print()

n = int(input("Enter number:"))
pattern10(n)
```

```python
# pattern 11
def pattern11(n):
    for i in range(1, n + 1):
        start = 1 if i % 2 == 1 else 0
        for j in range(i):
            print(start, end=" ")
            start = 1 - start
        print()

n = int(input())
pattern11(n)
```

```python
# pattern 12
def pattern12(n):
    spaces = 2 * (n - 1)

    for i in range(1, n + 1):
        for j in range(1, i + 1):
            print(j, end="")
        for j in range(spaces):
            print(" ", end="")
        for j in range(i, 0, -1):
            print(j, end="")
        print()
        spaces -= 2

n = int(input("Enter number:"))
pattern12(n)
```

```python
# pattern 13
def pattern13(n):
    num = 1
    for i in range(1, n + 1):
        for j in range(i):
            print(num, end=" ")
            num += 1
        print()

n = int(input("Enter number:"))
pattern13(n)
```

```python
# pattern 17
def pattern17(n):
    for i in range(n):
        for j in range(n - i - 1):
            print(" ", end="")

        ch = ord('A')
        breakpoint = (2 * i) // 2

        for j in range(1, 2 * i + 2):
            print(chr(ch), end="")
            if j <= breakpoint:
                ch += 1
            else:
                ch -= 1

        for j in range(n - i - 1):
            print(" ", end="")

        print()

n = int(input("Enter number:"))
pattern17(n)
```

```python
# pattern 18
def pattern18(n):
    for i in range(n):
        ch = chr(ord('A') + n - i - 1)
        for j in range(i + 1):
            print(chr(ord(ch) + j), end="")
        print()

n = int(input("Enter number:"))
pattern18(n)
```

```python
# pattern 19
def pattern19(n):
    for i in range(n):
        for j in range(n - i):
            print("*", end="")
        for j in range(2 * i):
            print(" ", end="")
        for j in range(n - i):
            print("*", end="")
        print()

    for i in range(n - 1, -1, -1):
        for j in range(n - i):
            print("*", end="")
        for j in range(2 * i):
            print(" ", end="")
        for j in range(n - i):
            print("*", end="")
        print()

n = int(input("Enter number:"))
pattern19(n)
```

```python
# Pattern 20
def pattern20(n):
    for i in range(n):
        for j in range(i):
            print("*", end="")
        for j in range(2 * (n - i)):
            print(" ", end="")
        for j in range(i):
            print("*", end="")
        print()

    for i in range(n - 2, -1, -1):
        for j in range(i):
            print("*", end="")
        for j in range(2 * (n - i)):
            print(" ", end="")
        for j in range(i):
            print("*", end="")
        print()

n = int(input("Enter number:"))
pattern20(n)
```

```python
# pattern 21
def pattern21(n):
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n - 1 or j == 0 or j == n - 1:
                print("*", end="")
            else:
                print(" ", end="")
        print()

n = int(input("Enter number:"))
pattern21(n)
```

```python
# Pattern 22
def pattern22(n):
    size = 2 * n - 1
    for i in range(size):
        for j in range(size):
            top = i
            left = j
            bottom = size - 1 - i
            right = size - 1 - j
            val = n - min(top, left, bottom, right)
            print(val, end=" ")
        print()

n = int(input("Enter number :"))
pattern22(n)
```
