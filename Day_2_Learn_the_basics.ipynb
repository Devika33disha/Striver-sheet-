# **Learn the Basics - Know basic maths**

```python
# 1 . Count all digits of a number
def countDigit(n):
  print(len(str(n)))
n = int(input("Enter number:"))
countDigit(n)
```

```python
#2. Reverse a number(positive)
def ReverseNumber(n):
  num = str(n)
  print(int(num[::-1]))
n = int(input("Enter number:"))
ReverseNumber(n)
```

```python
#3. Palindrome number
def ispalindrome(n):
  num = str(n)
  if num == num[::-1]:
    print(True)
  else:
    print(False)
n = int(input("Enter number:"))
ispalindrome(n)
```

```python
# 4. GCD of two numbers
def GCD(n1 , n2):
  while n2:
    n1,n2 = n2, n1%n2
  print(n1)
n1 = int(input("Enter first number:"))
n2 = int(input("Enter second number:"))
GCD(n1,n2)
#Another method using built in
# import math
# print(math.gcd(n1,n2))
```

```python
# 5. Check if the number is armstrong
def isArmstrong(n):
  num = str(n)
  sum = 0
  for i in range(len(num)):
    sum += int(num[i])** 3
  return sum == n

n = int(input("Enter number:"))
isArmstrong(n)
```

```python
# 6. Print all divisors
def divisors(n):
  lst = []
  for i in range(1,n+1):
    if n % i == 0:
      lst.append(i)
  print(lst)
n = int(input("Enter number:"))
divisors(n)
```

```python
#7.Check for prime number
def isPrime(n):
  lst = []
  for i in range(1,n+1):
    if n % i == 0:
      lst.append(i)
  return len(lst) == 2
n = int(input("Enter number:"))
isPrime(n)
```

# **Learn the basics - Learn Basic recursion**

```python
# 1. Print 1 to n using recursion
def printnumbers(n):
  if n < 1:
    return
  printnumbers(n-1)
  print(n)
n = int(input("Enter number:"))
printnumbers(n)
```

```python
# 2. Print n to 1 using recursion
def printnumbers(n):
  if n < 1:
    return
  print(n)
  printnumbers(n-1)
n = int(input("Enter number:"))
printnumbers(n)
```

```python
# 3. Sum of first N numbers
def NumbersSum(n):
  if n <= 1:
    return n
  return n+ NumbersSum(n-1)
n = int(input("Enter number:"))
NumbersSum(n)
```

```python
# 4. Factorial of a given number
def factorial(n):
  if n <= 1:
    return 1
  return n * factorial(n-1)
n = int(input("Enter Number:"))
factorial(n)
```

```python
# 5. Reverse an array
def reverse(arr,n):
  def helper(left,right):
      if left >= right:
          return
      arr[left], arr[right] = arr[right], arr[left]
      helper(left+1, right -1)
  helper(0,n-1)
n = int(input("Enter length of array:"))
arr = []
for i in range(n):
  arr.append(int(input(f"Enter element {i+1}:")))
reverse(arr,n)
print(arr)

```

```python
# 6. Check if string is palindrome or not
# s = str(input("Enter string:"))
# print(s==s[::-1])
# By recursion
def PalindromeCheck(s):
  def helper(left,right):
    if left >= right:
      return True
    if s[left] != s[right]:
      return False
    return helper(left +1 , right-1)
  return helper(0,len(s)-1)
s = input("Enter string :")
print(PalindromeCheck(s))
```

# **Learn the basics : learn basic hashing **

```python
# 1. Counting frequency of array elements
def countFrequency(nums):
  freq = {}
  for num in nums:
    freq[num] = freq.get(num,0) + 1
  return [[key,value] for key,value in freq.items()]
n = int(input("Enter length of array:"))
nums = []
for i in range(n):
  nums.append(int(input(f"Enter element {i+1}:")))
print(countFrequency(nums))

```

```python
# 2. Highest occuring element in an array
def mostFrequentelement(nums):
  freq = {}
  max_freq = 0
  ans = None
  for x in nums:
    freq[x] = freq.get(x,0) + 1
    if freq[x] > max_freq:
      max_freq = freq[x]
      ans = x
    elif freq[x] == max_freq and x < ans:
      ans = x
  return ans
n = int(input("Enter length of array:"))
nums = []
for i in range(n):
  nums.append(int(input(f"Enter element {i+1}:")))
print(mostFrequentelement(nums))
```
