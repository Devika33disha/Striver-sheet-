Selection sort
Step-by-step intuition
Imagine an unsorted list of numbers:

Step 1: Look at the whole list and find the smallest number.

Step 2: Swap this smallest number with the number at the first position.

Step 3: Now ignore the first position (it is sorted). From the remaining part of the list, again find the smallest number.

Step 4: Swap it with the number at the second position.

Step 5: Repeat this process: each time, find the smallest in the unsorted part and move it to the next position in the front.

```python
# selection sort
def SelectionSort(nums):
  n = len(nums)
  for i in range(n):
    min_index = i
    for j in range(i+1,n):
      if nums[j] < nums[min_index]:
        min_index = j
    nums[i],nums[min_index] = nums[min_index],nums[i]
  return nums
n = int(input("Length of nums:"))
nums = []
for i in range(n):
  nums.append(int(input(f"Enter element {i+1}: ")))
print(SelectionSort(nums))
```

# Bubble sort
Look at the array from left to right, comparing two neighbors at a time.

If the left element is bigger than the right one, swap them.

After one full pass, the largest element will have “bubbled up” to the end of the array.

Repeat this process several times; each pass pushes the next largest element to its correct position near the end.

```python
# Bubble sort
def BubbleSort(nums):
    n = len(nums)
    for i in range(n-1):
      for j in range(0,n-1-i):
        if nums[j] > nums[j+1]:
          nums[j],nums[j+1] = nums[j+1],nums[j]
    return nums
n = int(input("Enter length of nums:"))
nums = []
for i in range(n):
  nums.append(int(input(f"Enter element {i+1}: ")))
print(BubbleSort(nums))
```

# **Insertion sorting **
Imagine the array is a row of cards on the table.

The left part is treated as sorted, the right part is unsorted.

Start from the second element (index 1).

Take this element (call it key) and compare it with elements to its left.

Shift every bigger element one step to the right until you find the correct spot for key.

Put key in that spot.

Move to the next element and repeat.

```python
# Insertion sorting
def InsertionSort(nums):
  n = len(nums)
  for i in range(1,n):
    key = nums[i]
    j = i-1
    while j >= 0 and nums[j] > key :
      nums[j+1] = nums[j]
      j -= 1
    nums[j+1] = key
  return nums
n = int(input("Length of array: "))
nums = []
for i in range(n):
  nums.append(int(input(f"Enter element {i+1}: ")))
print(InsertionSort(nums))
```

# **Merge sorting**
Core idea in simple words
Step 1: Divide (split the array)
Keep splitting the array into two halves: left half and right half, again and again, until each piece has only one element. A single element is already “sorted” by itself.

Step 2: Conquer (merge in sorted order)
Now start merging those tiny sorted pieces back together:

Take two small sorted arrays.

Compare their first elements.

Pick the smaller one and put it into a new array.

Move forward in that array and keep comparing until all elements from both sides are taken.

The result is a bigger sorted array.
Repeat this merging step up the levels until you get one fully sorted array.



```python
# Merge sorting
def merge_sort(nums):
  if len(nums) <= 1:
    return nums
  mid = len(nums) // 2
  left = merge_sort(nums[:mid])
  right = merge_sort(nums[mid:])
  return merge(left,right)
def merge(left , right):
  i = j = 0
  merged = []
  while i < len(left) and j < len(right):
    if left[i] < right[j]:
      merged.append(left[i])
      i += 1
    else:
      merged.append(right[j])
      j += 1
  while i < len(left):
    merged.append(left[i])
    i += 1
  while j < len(right):
    merged.append(right[j])
    j += 1
  return merged
n = int(input("Length of array: "))
nums = []
for i in range(n):
  nums.append(int(input(f"Enter element {i+1}: ")))
print(merge_sort(nums))
```

```python
# Recursive bubble sort
def recursivebubblesort(nums):
  def helper(n):
      if n <= 1:
        return
      for i in range(n-1):
        if nums[i] > nums[i+1]:
          nums[i] , nums[i+1] = nums[i+1] , nums[i]
      helper(n-1)
  helper(len(nums))
  return nums
n = int(input("Length of array: "))
nums = []
for i in range(n):
  nums.append(int(input(f"Enter element {i+1}: ")))
print(recursivebubblesort(nums))
```

```python
# recursive insertion sort
def recursiveinsertionsort(nums):
  def firstsort(n):
    if n <= 1:
      return
    firstsort(n-1)
    key = nums[n-1]
    def insertpos(j):
      if j < 0 or nums[j] <= key:
          nums[j+1] = key
          return
      nums[j+1] = nums[j]
      insertpos(j-1)
    insertpos(n-2)
  firstsort(len(nums))
  return nums
n = int(input("Length of array: "))
nums = []
for i in range(n):
  nums.append(int(input(f"Enter element {i+1}: ")))
print(recursiveinsertionsort(nums))
```

```python
# Quick sort
def quicksort(nums):
  def helper(low , high):
    if low >= high:
      return
    pivoti = partition(low , high)
    helper(low , pivoti-1)
    helper(pivoti+1 , high)
  def partition(low,high):
    pivot = nums[high]
    i = low - 1
    for j in range(low , high):
      if nums[j] <= pivot:
         i += 1
         nums[i] , nums[j] = nums[j] , nums[i]
    nums[i+1] , nums[high] = nums[high] , nums[i+1]
    return i + 1
  helper(0,len(nums)-1)
  return nums
n = int(input("Length of array: "))
nums = []
for i in range(n):
  nums.append(int(input(f"Enter element {i+1}: ")))
print(quicksort(nums))
```
